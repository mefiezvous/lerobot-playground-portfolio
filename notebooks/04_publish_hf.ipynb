# Publish to HuggingFace Hub

Push the dataset and trained model checkpoints to the Hub.

**Targets:**
- Dataset  : `mefiezvous/cube-reach-v1-dataset`
- ACT model : `mefiezvous/cube-reach-v1-act`
- Diffusion model: `mefiezvous/cube-reach-v1-diffusion`

> **Important:** Only push data from the PUBLIC layer. Never reference private checkpoints here.

```python
# Cell 1 — Authenticate to HuggingFace
from huggingface_hub import login

# Paste your HF token with write access, or use HF_TOKEN env var
login()  # interactive login; use login(token=os.environ["HF_TOKEN"]) on Kaggle
```

```python
# Cell 2 — Push LeRobotDataset to Hub
# Requires the dataset to have been collected locally first (see 01_quickstart.ipynb)
from lerobot.common.datasets.lerobot_dataset import LeRobotDataset  # type: ignore[import]

DATASET_REPO_ID = "mefiezvous/cube-reach-v1-dataset"
LOCAL_DATASET_PATH = "../data/cube_reach_v1_dataset"  # adjust to your local path

dataset = LeRobotDataset(DATASET_REPO_ID, root=LOCAL_DATASET_PATH)
dataset.push_to_hub(
    repo_id=DATASET_REPO_ID,
    private=False,
)
print(f"Dataset pushed to https://huggingface.co/datasets/{DATASET_REPO_ID}")
```

```python
# Cell 3 — Push ACT model checkpoint
from huggingface_hub import upload_folder

ACT_REPO_ID = "mefiezvous/cube-reach-v1-act"
ACT_CHECKPOINT_DIR = "../checkpoints/act/"  # directory with checkpoint_XXXXXXXX.ckpt

upload_folder(
    repo_id=ACT_REPO_ID,
    folder_path=ACT_CHECKPOINT_DIR,
    repo_type="model",
    commit_message="Upload ACT checkpoint from training run",
    ignore_patterns=["*.tmp", "__pycache__"],
)
print(f"ACT model pushed to https://huggingface.co/{ACT_REPO_ID}")
```

```python
# Cell 4 — Push Diffusion model checkpoint
DIFF_REPO_ID = "mefiezvous/cube-reach-v1-diffusion"
DIFF_CHECKPOINT_DIR = "../checkpoints/diffusion/"

upload_folder(
    repo_id=DIFF_REPO_ID,
    folder_path=DIFF_CHECKPOINT_DIR,
    repo_type="model",
    commit_message="Upload Diffusion checkpoint from training run",
    ignore_patterns=["*.tmp", "__pycache__"],
)
print(f"Diffusion model pushed to https://huggingface.co/{DIFF_REPO_ID}")
```

```python
# Cell 5 — Verify published artifacts
from huggingface_hub import HfApi

api = HfApi()
for repo_id, repo_type in [
    ("mefiezvous/cube-reach-v1-dataset", "dataset"),
    ("mefiezvous/cube-reach-v1-act", "model"),
    ("mefiezvous/cube-reach-v1-diffusion", "model"),
]:
    info = api.repo_info(repo_id=repo_id, repo_type=repo_type)
    print(f"{repo_type:8s} {repo_id:50s} ✓ (id={info.id})")
```
