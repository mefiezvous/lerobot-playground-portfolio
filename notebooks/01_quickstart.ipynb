Run scripted demos in the `cube_reach_v1` environment and inspect the collected dataset.

**Requirements:** CPU only · no GPU needed · runs in ~30 s

```python
# Cell 1 — Environment setup
# Install the package in editable mode (skip if already installed)
%pip install -e .. -q
```

```python
# Cell 2 — Collect 5 scripted episodes
from playground.data.pipeline import DemoCollector, ScriptedPolicy
from playground.envs.cube_reach_v1 import CubeReachV1

env = CubeReachV1()
policy = ScriptedPolicy(env)
collector = DemoCollector(
    env=env,
    policy=policy,
    repo_id="local/quickstart-demo",
    n_episodes=5,
)

dataset = collector.collect()
print(f"Collected {len(dataset)} episodes")
```

```python
# Cell 3 — Inspect results as a DataFrame
import pandas as pd

records = []
for ep_idx in range(len(dataset)):
    ep = dataset[ep_idx]
    records.append({
        "episode": ep_idx,
        "n_steps": len(ep["action"]) if hasattr(ep["action"], "__len__") else 1,
        "reward": float(ep.get("reward", 0.0)),
    })

df = pd.DataFrame(records)
print(df.to_string(index=False))
print(f"\nMean reward: {df['reward'].mean():.3f}")
```
